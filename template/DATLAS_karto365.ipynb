## PROCESS a pd DataFrame ```pdf_tables_interest``` 
In this section:
1. ```try_get_pdf_table()``` tries to extract all the info from the tables specified in ```pdf_tables_interest```
2. Two approaches are attempted to get the information from the table
   1. ```Approach 1```, using method ```urlopen(url).read().decode()```
   2. ```Approach 2```, using method ```requests.get(url).json()```
3. Then, if either Approach is successful, a json dictionary is returned, used to build a pd DataFrame using ```normalize_pdf_columns()``` with
      1. Column ```col_features:str = 'features'```, and
      2. Column ```col_attribs:str = 'attributes'```
4. The initial pd DataFrame ```pdf_tables_interest``` is enriched with:
   1. Column ```shape``` with the size rows/columns of the table extracted
   2. Column ```approach``` storing the approach used (either 1-urlopen(url).read().decode(), or 2.requests.get(url).json()
   3. Column ```table_info``` contains the information of the table extracted

````python
l_shape = []
l_approach = []
l_table_info = []
for i,table in enumerate(list(pdf_tables_interest.title)):
    
    print(f'table={table}[{i}/{pdf_tables_interest.shape[0]}]')
    try:
        [pdf_table, table_approach, dict_table_info] = web_functions.try_get_pdf_table(table=table)
        table_shape = pdf_table.shape
    except:
        table_shape = None
        table_approach = None
        dict_table_info = None
        print('Error!')
    finally:
        # print(f'shape={l_shape}')
        # print(f'approach={l_approach}')
        l_shape.append(table_shape)
        l_approach.append(table_approach)
        l_table_info.append(dict_table_info)
        print()
````
